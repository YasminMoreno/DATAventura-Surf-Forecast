```python
#%pip install geopy
from geopy.geocoders import Nominatim

import time

import math
```

```python
geo = Nominatim(user_agent = 'localizador')
```

```python
loc = geo.geocode('Playa del Cotillo')
print(loc)

la = loc.latitude 

lo = loc.longitude

print(la)

print(lo)
```

```python
#%pip install requests
import requests as req

#%pip install arrow

import arrow
#%pip install pandas

import pandas as pd

import numpy as np

import json

from urllib.request import urlopen
from pandas import json_normalize
```

```python
#start = arrow.now().floor('day')

start = arrow.get('2022-09-28 01:00')

# Get last hour of today
end = arrow.now().ceil('day')


response = req.get(
  'https://api.stormglass.io/v2/weather/point',
  params={
    'lat': la,
    'lng': lo,
    'params': ','.join(['airTemperature', 'humidity', 'precipitation', 'windSpeed', 'windDirection', 'swellHeight', 'swellDirection', 
                        'swellPeriod','secondarySwellHeight', 'secondarySwellDirection', 'secondarySwellPeriod', 
                        'seaLevel', 'currentDirection', 'currentSpeed']),
    'end': end.to('UTC').timestamp(),
    'start': start.to('UTC').timestamp(), # Convert to UTC timestamp
  },
  headers={
    'Authorization': '530edd60-3fe6-11ed-aa92-0242ac130002-530ede1e-3fe6-11ed-aa92-0242ac130002'
  }
)

# Do something with response data.
json_data = response.json()
```
