# Formulas para agregar columnas

```python
niveles = []


for i in range(len(surf_proyect_final.altura_del_oleaje)):
    if surf_proyect_final.altura_del_oleaje[i] <= 0.5 and surf_proyect_final.periodo_del_oleaje[i] >= 7 and surf_proyect_final.periodo_del_oleaje[i] <= 9:
        variable ='primera_vez'     
    elif surf_proyect_final.altura_del_oleaje[i] <= 1.5 and surf_proyect_final.periodo_del_oleaje[i] >= 7 and surf_proyect_final.periodo_del_oleaje[i] <= 9:
        variable = 'principiante'
    elif surf_proyect_final.altura_del_oleaje[i] <= 3 and surf_proyect_final.periodo_del_oleaje[i] <= 12:
        variable = 'intermedio'
    elif surf_proyect_final.altura_del_oleaje[i] <= 4 and surf_proyect_final.periodo_del_oleaje[i] <= 15:
        variable = 'avanzado'
    else:
        variable = 'profesional'
        
    niveles.append(variable)
    

niveles[-5:]
```

```python
surf_proyect_final['niveles']=niveles
```

```python
surf_proyect_final.head()
```

```python
letra_viento = []

for i in range(len(surf_proyect_final.direccion_del_viento)):
    # Rumbos Francos
    if surf_proyect_final.direccion_del_viento[i] == 0 or surf_proyect_final.direccion_del_viento[i] == 360:
        rumbo = 'N'
    elif surf_proyect_final.direccion_del_viento[i] == 90:
        rumbo = 'E'
    elif surf_proyect_final.direccion_del_viento[i] == 180:
        rumbo = 'S'
    elif surf_proyect_final.direccion_del_viento[i] == 270:
        rumbo = 'W'
    #Identificar los cuadrantes.
    #Cuadrante I
    elif surf_proyect_final.direccion_del_viento[i] > 0 and surf_proyect_final.direccion_del_viento[i] < 90:
        rumbo = 'N ' + 'E'
    #Cuadrante II
    elif surf_proyect_final.direccion_del_viento[i] > 90 and surf_proyect_final.direccion_del_viento[i] < 180:
        rumbo = 'S ' + 'E'
    #Cuadrante III
    elif surf_proyect_final.direccion_del_viento[i] > 180 and surf_proyect_final.direccion_del_viento[i] < 270:
        rumbo = 'S ' + 'W'
    #Cuadrante IV
    elif surf_proyect_final.direccion_del_viento[i] > 270 and surf_proyect_final.direccion_del_viento[i] < 360:
        rumbo = 'N ' + 'W'
    else:
        rumbo = 'Valor Incorrecto'
        
        
    letra_viento.append(rumbo)
    
letra_viento
```

```python
surf_proyect_final['coordenadas_viento']=letra_viento

surf_proyect_final
```
